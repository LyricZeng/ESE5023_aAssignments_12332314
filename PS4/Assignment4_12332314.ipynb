```python
import pandas as pd
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.transforms import offset_copy
%matplotlib inline
```

# 1.Global Earthquakes

```python
ds = pd.read_csv("usgs_earthquakes.csv")
df = ds.sort_values(by='mag',ascending=False).head(50)
df = df.reset_index(drop=True)
df
```

```python
fig = plt.figure(figsize=(10, 5))#尺寸10*5
#在图形中添加一个子图，并将其投影设置为Robinson，中心经度为180度
ax = fig.add_subplot(1, 1, 1, projection=ccrs.Robinson(central_longitude=180,globe=None))
ax.set_global()#设置坐标轴的全球范围
ax.stock_img()#添加背景图
ax.set_title('Top 50 Earthquakes of 2014')#标题
#在地图上绘制地震数据，使用散点图，点的位置由经纬度决定，点的颜色表示地震的震级，颜色映射用Reds，点的边缘颜色为黑色。
eq = plt.scatter(df['longitude'].values,df['latitude'].values, marker='o', c=df['mag'], cmap='Reds', edgecolors='k', transform=ccrs.PlateCarree())
# 在图形中添加一个颜色条，颜色条的颜色映射和散点图的颜色映射是一样的
cb = fig.colorbar(eq,cax=fig.add_axes([0.95,0.25,0.01,0.5]),shrink=0.7)
cb.set_label('magnitude')#设置颜色条标签：地震级别
```

# 2.Explore a netCDF dataset

## 2.1

```python
data = xr.open_dataset("MERRA2.statM_2d_pct_Nx.nc4")#MERRA2是一个全球大气再分析数据集
fig = plt.figure(figsize=(16, 8))#设置大小为宽16英寸、高8英寸的figure
#添加一个子图到figure对象中，并将其投影设置为平面圆盘投影（Plate Carree）
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.set_global()#全球范围
ax.stock_img()#全球底图
ax.coastlines()#绘制海岸线
#添加国家边界特征，但不显示面填充颜色，只显示黑色边缘。
ax.add_feature(cfeature.NaturalEarthFeature(category='cultural',
                                           name='admin_0_countries',
                                           scale='110m',
                                           facecolor='none',
                                           edgecolor='black',
                                           linewidth=0.5))
data['T2MMEAN'].plot(transform=ccrs.PlateCarree(),cmap='Spectral')#从打开的数据集中选择’T2MMEAN’变量，并在地图上绘制，使用’Spectral’颜色映射
x, y =  [73,73,135,135] , [3.8,54,54,3.8]#经纬度范围
ax.plot(x, y, marker='o',color='blue', transform=ccrs.PlateCarree())#绘制多边形，用蓝色标记
ax.fill(x, y, color='grey', transform=ccrs.PlateCarree(), alpha=0.4)#灰色填充多边形，透明度为0.4

ax.plot(116.25,39.54, marker='o', color='red', markersize=12,
            alpha=0.7, transform=ccrs.PlateCarree())#标记北京的位置
#获取地理变换对象
geodetic_transform = ccrs.Geodetic()._as_mpl_transform(ax)
# 创建文本变换对象，将文本绘制在地图上的特定位置。
text_transform = offset_copy(geodetic_transform, units='dots', x=-10)
#北京，文本背景框为棕色，设置透明度及边框样式
ax.text(116.25,39.54, 'BeiJing',
            verticalalignment='center', horizontalalignment='right',
            transform=text_transform,
            bbox=dict(facecolor='sandybrown', alpha=0.5, boxstyle='round'))
ax.gridlines(draw_labels=True,linestyle=":",linewidth=1.5,color='k')#绘制经纬度网格，并显示标签
```

## 2.2

```python
plt.figure(figsize=(10,10),dpi=100)
central_lon, central_lat = 116.25,39.54 #北京
#创建一个投影对象，这个投影是Lambert共形投影，并且以北京的经纬度为中心。
proj = ccrs.LambertConformal(central_lon,central_lat)
#创建一个坐标轴对象，并将投影对象应用到这个坐标轴上。
ax = plt.axes(projection=proj)
ax.set_global()#坐标轴的全球范围
ax.stock_img()#默认的背景图（蓝色海洋）
ax.coastlines()#添加海岸线
ax.gridlines(draw_labels=True,linestyle=":",linewidth=1.5,color='k')#添加经纬度网格线，显示网格标签
ax.add_feature(cfeature.NaturalEarthFeature(category='cultural',
                                           name='admin_0_countries',
                                           scale='110m',
                                           facecolor='none',
                                           edgecolor='black',
                                           linewidth=0.5))#添加国家边界，设置面颜色：无，显示边框颜色为黑色
extent = [central_lon-30,central_lon+18,central_lat-35.74,central_lat+14.46]#显示范围
data['T2MMEAN'].plot(transform=ccrs.PlateCarree(),cmap='Spectral')#绘制气温数据
ax.set_extent(extent)#设置地图显示范围
ax.add_feature(cfeature.LAND, edgecolor='black')#添加陆地图层，边缘颜色为黑色
ax.plot(114.06,22.54, marker='o', color='red', markersize=12,
            alpha=0.7, transform=ccrs.PlateCarree())#标记红色点

geodetic_transform = ccrs.Geodetic()._as_mpl_transform(ax)#获取地理变换对象
#创建一个文本变换对象，用于将文本向左偏移10个单位。
text_transform = offset_copy(geodetic_transform, units='dots', x=-10)
#创建一个文本变换对象，用于将文本向右偏移10个单位。
text_transform1 = offset_copy(geodetic_transform, units='dots', x=+10)
#在标记点上添加文本，显示该点的气温值。
ax.text(114.06,22.54,data.T2MMEAN.sel(lon='114.0',lat='22.5', method='nearest').values,
            verticalalignment='center', horizontalalignment='left',
            transform=text_transform1,
            bbox=dict(facecolor='sandybrown', alpha=0.5, boxstyle='round'))
#标记点为深圳
ax.text(114.06,22.54,'shenzhen',
            verticalalignment='center', horizontalalignment='right',
            transform=text_transform,
            bbox=dict(facecolor='sandybrown', alpha=0.5, boxstyle='round'))
```

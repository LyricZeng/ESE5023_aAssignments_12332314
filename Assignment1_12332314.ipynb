## PS1_3 Pascal triangle

```python
#3.Pascal triangle
def generate_pascals_triangle(n):    
     triangle = [[1]]    
     for i in range(1, n):    
         row = [1]    
         for j in range(1, i):    
             row.append(triangle[i-1][j-1] + triangle[i-1][j])    
         row.append(1)    
         triangle.append(row)    
     return triangle

n_values = [100, 200]

for n in n_values:  
    pascals_triangle = generate_pascals_triangle(n)    
    for row in pascals_triangle:    
        print(row)
```

```python
#5.Dynamic programming
#5.1
from tqdm import *#进度条
import itertools
import numpy as np  
A = np.random.randint(1, 100)
def Find_expression(A: int, nums: str) -> list:  
    opra = ['+', '-', '']  
    def find_all(nums: str) -> list:  
        if len(nums) == 1:  
            return [nums]  
        else:  
            return [nums[0] + j + i for i in find_all(nums[1:]) for j in opra]        
    # 生成所有可能的组合  
    all_combinations = list(itertools.product(opra, repeat=len(nums)))   
    # 过滤出满足条件的表达式  
    results = [k + '=' + str(A) for k in find_all(nums) if eval(k) == A]       
    return results
if __name__ == "__main__":  
    results = {}
    for a in tqdm(range(1,101)):
        result = Find_expression(a, "123456789")  
        results[a] = result
print(results)
```
